```python
states = data['Provider State'].unique()
print(states)
```

```python
discharge_data = []
for state in states:
    temp = data[data['Provider State'] == state][' Total Discharges '].values
    discharge_data.append(temp)
#print(discharge_data)    
    
    
```

```python
print(discharge_data[0])
temp = 0
for i in discharge_data[0]:
    #print(i)
    temp += i
#print(temp)
dataset_count = []
for i in range(0,51):
    for j in discharge_data[i]:
        temp+=j
    dataset_count.append(temp)
print(dataset_count)        
        
```

```python
#print(discharge_data)
#print(len(discharge_data))
#print(discharge_data[0])
##temp = 0
#for i in discharge_data:
 #   temp += i
#print(temp)   

plt.figure(figsize = (15,8))
plt.title('Total Discharge in hospitals in various states')
ax = plt.subplot()
ax.set_xticks(state_numbers)
ax.set_xticklabels(states)
plt.bar(state_numbers,dataset_count)
plt.plot(dataset_count,color='green')
```

```python
avg_covered_payments = []
for state in states:
    temp = data[data['Provider State'] == state][' Average Covered Charges '].values
    avg_covered_payments.append(temp)
print(avg_covered_payments[50])    
print(len(avg_covered_payments[0]))
```

```python
len_avg_covered_payments = []

for i in range(0,51):
    len_avg_covered_payments.append(len(avg_covered_payments[i]))
print(len_avg_covered_payments)    
```

```python
total_state_covered_payments = []

for i in range(0,51):
    temp = 0
    for j in avg_covered_payments[i]:
        temp+=j
    total_state_covered_payments.append(temp)
print(total_state_covered_payments)        
```

```python
final_covered_payment = []
j = 0
for i in total_state_covered_payments: #and for j in range(0,51):
    
        temp = i / len_avg_covered_payments[j]     
        print(temp)
        j+=1
        final_covered_payment.append(temp)
    
#print(final_covered_payment)    
#print(len(final_covered_payment))
```

```python
avg_total_payments = []
for state in states:
    temp = data[data['Provider State'] == state][' Average Total Payments '].values
    avg_total_payments.append(temp)
#print(avg_total_payments)  

total_state_payments = []

for i in range(0,51):
    temp = 0
    for j in avg_total_payments[i]:
        temp+=j
    total_state_payments.append(temp)
#print(total_state_payments)        

final_total_payment = []
j = 0
for i in total_state_payments: #and for j in range(0,51):
    
        temp = i / len_avg_covered_payments[j]     
        print(temp)
        j+=1
        final_total_payment.append(temp)
```

```python
avg_medicare_payments = []
for state in states:
    temp = data[data['Provider State'] == state]['Average Medicare Payments'].values
    avg_medicare_payments.append(temp)
#print(avg_total_payments)  

total_medicare_payments = []

for i in range(0,51):
    temp = 0
    for j in avg_medicare_payments[i]:
        temp+=j
    total_medicare_payments.append(temp)
#print(total_state_payments)        

final_medicare_payment = []
j = 0
for i in total_medicare_payments: #and for j in range(0,51):
    
        temp = i / len_avg_covered_payments[j]     
        print(temp)
        j+=1
        final_medicare_payment.append(temp)
```

```python
plt.figure(figsize = (18,8))
ax1 = plt.subplot()
ax1.set_xticks(state_numbers)
ax1.set_xticklabels(states)
#plt.legend(['Average Covered Charges','Average Total Payments'])
plt.plot(final_covered_payment)
plt.plot(final_total_payment)
plt.plot(final_medicare_payment)
plt.legend(['Average Covered Charges','Average Total Payments','Average Medicare Payments'])
plt.xlabel('US States')
plt.ylabel('Amount in US dollars')
plt.title('Average healthcare amounts in various US states')
```
